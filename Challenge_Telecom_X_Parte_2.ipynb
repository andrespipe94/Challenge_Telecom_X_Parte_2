##**Eliminación de columnas irrelevantes**

```python
datos.columns
```

```python
datos.drop(columns=['Cliente_ID','Genero','Pareja','Dependientes','Servicio_telefonico','Varias_lineas','Servicio_internet','Copia_seguridad_en_linea','Proteccion_dispositivo','Televisión_por_cable','Streaming_peliculas','Factura_en_linea','Cuentas_diarias', 'N_servicios'], axis=1, inplace=True)
datos.info()
```

##**Separación de Datos**

```python
X = datos.drop('Abandono', axis=True)
y = datos.Abandono

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, stratify=y, random_state=123)
```

## **Encoding**

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
```

```python
columnas = X.columns
```

```python
one_hot = make_column_transformer((OneHotEncoder(drop='if_binary', handle_unknown='ignore'),
                                   ['Tipo_servicio_internet', 'Tipo_contrato', 'Forma_pago']),
                                  remainder='passthrough',
                                  sparse_threshold=0,
                                  force_int_remainder_cols=False)
```

```python
one_hot.fit(X_train)
```

```python
X_train_transf = one_hot.transform(X_train)
X_test_transf = one_hot.transform(X_test)
```

##**Verificación de la Proporción de Cancelación (Churn)**

```python
datos['Abandono'].value_counts(normalize=True).round(2)
```

##**Balanceo de Clases**

```python
from imblearn.over_sampling import SMOTE
```

```python
oversampling = SMOTE(random_state=123)
X_train_bal, y_train_bal = oversampling.fit_resample(X_train_transf,y_train)
y_train_bal.value_counts(normalize=True)
```

##**Normalizando los datos**

La **normalización** de los datos garantiza que todas las variables contribuyan en igualdad de condiciones al proceso de modelado. En modelos basados en distancia, como **KNN**, la magnitud de las variables influye directamente en el cálculo de la distancia entre observaciones. Si las variables no están escaladas, aquellas con mayor rango numérico pueden dominar la métrica de distancia e introducir un sesgo en la clasificación.

Al aplicar una normalización (por ejemplo, mediante **MinMaxScaler**), todas las variables se reescalan a un mismo rango —habitualmente entre 0 y 1—, lo que evita que aquellas con valores numéricos más altos dominen el proceso de aprendizaje. De esta manera, la magnitud original de cada variable no determina su influencia en modelos basados en distancia (como **KNN**) o en optimización de parámetros (como la **regresión logística** o las **redes neuronales**).

En contraste, los **modelos basados en árboles no requieren normalización**, ya que realizan particiones mediante umbrales y no dependen de métricas de distancia.

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
normalizacion = MinMaxScaler()
X_train_bal_normalizado = normalizacion.fit_transform(X_train_bal)
X_test_normalizado = normalizacion.transform(X_test_transf)
```

```python
pd.DataFrame(X_train_bal_normalizado)
```

#**Creación de Modelos**

##**Baseline**

```python
from sklearn.dummy import DummyClassifier
```

```python
dummy = DummyClassifier(strategy='stratified')
```

```python
dummy.fit(X_train_transf, y_train)
```

```python
dummy.score(X_test, y_test)
```

##**Árbol de decisión**

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
modelo_arbol = DecisionTreeClassifier(max_depth=4, min_samples_split=2, min_samples_leaf=1, random_state=123)
modelo_arbol.fit(X_train_bal, y_train_bal)
```

```python
modelo_arbol.score(X_test_transf, y_test)
```

##**KNN**

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=15, weights='distance')
```

```python
knn.fit(X_train_bal_normalizado, y_train_bal)
```

```python
knn.score(X_test_normalizado, y_test)
```

##**Evaluación de los Modelos**

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score

y_dummy_predict = dummy.predict(X_test_transf)
y_arbol_predict = modelo_arbol.predict(X_test_transf)
y_knn_predict = knn.predict(X_test_normalizado)


resultados = (pd.DataFrame({
    'Modelo' : ['Dummy', 'Decision Tree', 'KNN'],
    'Accuracy':[
        accuracy_score(y_test, y_dummy_predict),
        accuracy_score(y_test, y_arbol_predict),
        accuracy_score(y_test, y_knn_predict)
    ],
    'Recall' : [
        recall_score(y_test, y_dummy_predict),
        recall_score(y_test, y_arbol_predict),
        recall_score(y_test, y_knn_predict)
    ],
    'Precision' : [
        precision_score(y_test, y_dummy_predict),
        precision_score(y_test, y_arbol_predict),
        precision_score(y_test, y_knn_predict),
    ],
    'F1': [
        f1_score(y_test, y_dummy_predict),
        f1_score(y_test, y_arbol_predict),
        f1_score(y_test, y_knn_predict)
    ]
})).round(2)


resultados
```

###**Matriz de confusión**

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_dummy_predict, display_labels=["No Abandona", "Abandona"])
plt.title('Matriz de Confusiín - Dummy')
plt.show()
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_arbol_predict, display_labels=["No Abandona", "Abandona"])

plt.title("Matriz de Confusión - Decision Tree")
plt.show()
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_knn_predict, display_labels=["No Abandona", "Abandona"])

plt.title("Matriz de Confusión - KNN")
plt.show()
```

#**Análisis de la Importancia de las Variables**

##**KNN**

```python
from sklearn.inspection import permutation_importance

resultados = permutation_importance(
    knn,
    X_test_normalizado,
    y_test,
    n_repeats=10,
    random_state=123,
    scoring='recall'
)

nombre_columnas = one_hot.get_feature_names_out()

importancia_knn = pd.Series(resultados.importances_mean, index=nombre_columnas)
importancia_knn.sort_values(ascending=False)
```

##**Modelo Árbol**

```python
importancia_arbol = pd.Series(
    modelo_arbol.feature_importances_,
    index=nombre_columnas
)

importancia_arbol.sort_values(ascending=False)
```

```python
resultados_arbol = permutation_importance(
    modelo_arbol,
    X_test_transf,
    y_test,
    n_repeats=10,
    random_state=123,
    scoring='recall'
)

importancia_arbol = pd.Series(
    resultados_arbol.importances_mean,
    index=nombre_columnas
)

importancia_arbol.sort_values(ascending=False)
```

##**Principales factores que afectan la cancelación**

1. Contrato mensual (Month-to-month)

Es el principal predictor de churn.

Los clientes sin compromiso de permanencia presentan una probabilidad significativamente mayor de cancelar. Esto sugiere que los contratos de mayor duración funcionan como una barrera de salida.

2. Pago mediante Electronic Check

Los medios de pagos poco automatizados están asociados a mayor cancelación.

3. Servicio de Internet por Fibra Óptica

Muestra asociación positiva con churn.

Este resultado permite identificar a la fibra óptica como un segmento de alto riesgo, compuesto por clientes más exigentes, sensibles a problemas de calidad del servicio y a precios más elevados.

4. Baja antigüedad (Meses_permanencia)

Clientes con menor permanencia presentan mayor probabilidad de cancelar.

El riesgo de abandono es mayor en los primeros meses del servicio.

##**Estrategias de Retención Propuestas**

1. **Incentivar contratos de largo plazo**

Incentivar la migración hacia contratos a plazo mediante descuentos exclusivos y paquetes con servicios adicionales bonificados, generando barreras de salida amigables sin afectar la percepción del cliente.

2. **Programa de fidelización temprana**

Diseñar un programa que contemple llamadas proactivas los primeros 3-6 meses, encuestas de satisfacción sobre la calidad del servicio y beneficios exclusivos iniciales, con el objetivo de que el cliente supere la zona crítica de permanencia. Este programa debería focalizarse especialmente en clientes con contratos mes a mes, servicio de fibra óptica y formas de pago no automatizadas.

3. **Revisión del segmento fibra óptica**

Analizar competitividad de precios, revisar la calidad percibida del servicio y ofrecer paquetes diferenciales con servicios complementarios, como soporte técnico o seguridad en línea, con el fin de mejorar la experiencia del usuario.

4. **Promoción de pagos automáticos**

Promover la adopción de medios de pago automatizados mediante descuentos, beneficios exclusivos o promociones especiales, reduciendo la fricción operativa y desincentivando el abandono.

##**Conclusión Final**

El modelo de **Árbol de Decisión** demostró ser el más adecuado para el objetivo del negocio, ya que maximiza la detección de clientes que cancelan (recall 0.80).

El análisis de importancia de variables indica que:

La **modalidad de contrato** es el factor determinante.

La **antigüedad** del cliente es clave.

El **método de pago** y el **tipo de servicio** influyen significativamente.

Las estrategias de retención deben enfocarse principalmente en clientes con contrato mensual, baja permanencia y pago mediante electronic check.
